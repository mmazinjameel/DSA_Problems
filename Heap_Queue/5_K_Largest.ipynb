```python
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        nums = [-s for s in nums]          # Convert to min-heap using negative values
        heapq.heapify(nums)                # Heapify the list

        for i in range(k - 1):             # Pop k-1 elements to get to the k-th largest
            heapq.heappop(nums)

        return -heapq.heappop(nums)        # Return the k-th largest (invert the sign)
```
