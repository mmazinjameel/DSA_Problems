```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:

        if len(s) == 0:
            return 0

        word_set = set()
        max_sub_len = 0

        i = 0
        for j in range(len(s)):
            while s[j] in word_set:
                word_set.remove(s[i])
                i+=1
            
            word_set.add(s[j])
            max_sub_len = max(max_sub_len, j-i+1)

        return max_sub_len



```
