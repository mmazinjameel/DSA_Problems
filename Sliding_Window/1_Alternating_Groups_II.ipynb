```python
class Solution:
    def numberOfAlternatingGroups(self, colors: List[int], k: int) -> int:
        n = len(colors)
        count = 0
        streak = 1   # length of current alternating sequence

        for i in range(1, n + k):
            curr = colors[i % n]
            prev = colors[(i - 1) % n]

            if curr != prev:
                streak += 1
            else:
                streak = 1

            if streak >= k:
                start = i - k + 1
                if start < n:   # ensure start is within original array
                    count += 1
        
        return count
```
