```python
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        v1 = [0] * 26

        for s in s1:
            v1[ord(s) - ord('a')] += 1

        l = 0
        v2 = [0] * 26

        for r in range(len(s2)):
            v2[ord(s2[r]) - ord('a')] += 1

            if r - l + 1 == len(s1):
                if v1 == v2:
                    return True
                else:
                    v2[ord(s2[l]) - ord('a')] -= 1
                    l += 1

        return False
```
