```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        
        def canFinish(piles, h, k):
            hours = 0
            for pile in piles:
                hours += (pile + k - 1) // k  # ceiling division
            return hours <= h
        
        left = 1
        right = max(piles)

        while left < right:
            mid = (left + right) // 2
            if canFinish(piles, h, mid):
                right = mid
            else:
                left = mid + 1

        return left
```
