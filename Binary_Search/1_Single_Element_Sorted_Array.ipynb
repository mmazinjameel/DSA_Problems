```python
class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        
        lo, hi = 0, len(nums)-1
        while lo < hi:
            mid = (lo + hi) // 2

            if nums[mid] != nums[mid+1] and nums[mid] != nums[mid-1]:
                return nums[mid]

            # Check left side pair
            if nums[mid] == nums[mid-1]:
                if (mid - 1) % 2 == 0:
                    lo = mid + 1
                else:
                    hi = mid - 1

            # Check right side pair
            elif nums[mid] == nums[mid+1]:
                if mid % 2 == 0:
                    lo = mid + 2
                else:
                    hi = mid - 1

        return nums[lo]
```
