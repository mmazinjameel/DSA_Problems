```python
class Solution:
    def shipWithinDays(self, weights: List[int], days: int) -> int:


        def canShip(weights, days, capacity):
            days_used = 1
            current = 0

            for w in weights:
                if current + w > capacity:
                    days_used += 1
                    current = 0
                current += w

            return days_used <= days

        left = max(weights)
        right = sum(weights)

        while left < right:
            mid = (left + right) // 2
            if canShip(weights, days, mid):
                right = mid
            else:
                left = mid + 1

        return left
```
