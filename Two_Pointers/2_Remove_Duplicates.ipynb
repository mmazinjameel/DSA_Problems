```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums) <= 2:
            return len(nums)

        k = 2  # first two elements are always allowed

        for i in range(2, len(nums)):
            if nums[i] != nums[k - 2]:
                nums[k] = nums[i]
                k += 1

        return k
```

### Another Solution

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i = 0
        num = nums[0]
        count = 0

        while i<len(nums):

            if nums[i]==num:
                count+=1

            if nums[i]!=num:
                num = nums[i]
                count=1

            if count>=3:
                nums.pop(i)
                continue

            i+=1

        return len(nums)
```
