```python
class Solution:
    def countFairPairs(self, nums, lower, upper):
        nums.sort()

        def count_less_equal(target):
            left = 0
            right = len(nums) - 1
            count = 0

            while left < right:
                if nums[left] + nums[right] <= target:
                    count += (right - left)
                    left += 1
                else:
                    right -= 1

            return count

        return count_less_equal(upper) - count_less_equal(lower - 1)
```
