```python
class Solution:
    def countSubIslands(self, grid1: List[List[int]], grid2: List[List[int]]) -> int:
        
        rows, cols = len(grid1), len(grid1[0])
        count = 0

        def dfs(i,j):
            if i<0 or i>=rows or j < 0 or j >= cols or grid2[i][j] != 1:
                return True

            # mark it as visited
            grid2[i][j] = -1  # mark as visited
            is_sub = grid1[i][j] == 1

            # AND across all directions
            up = dfs(i+1, j)
            down = dfs(i-1, j)
            left = dfs(i, j+1)
            right = dfs(i, j-1)

            return is_sub and up and down and left and right

        for i in range(rows):
            for j in range(cols):
                if grid2[i][j]==1:
                    count += int(dfs(i, j))
                    
        return count
        
```
