```python
class Solution:
    def isAlienSorted(self, words: List[str], order: str) -> bool:
        char_index = {char: index for index,char in enumerate(order)}

        for i in range(len(words)-1):
            w1 = words[i]
            w2 = words[i+1]

            j = 0

            while j < len(w1) and j < len(w2):
                
                if char_index[w1[j]] < char_index[w2[j]]:   # Found the first difference and it is smaller
                    break 

                elif char_index[w1[j]] > char_index[w2[j]]:
                    return False

                j += 1

            else:
                if len(w1) > len(w2):
                    return False

        return True  
```
