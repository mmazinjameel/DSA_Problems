```python
class Solution:
    def minReorder(self, n: int, connections: List[List[int]]) -> int:

        graph = defaultdict(list)

        for a, b in connections:
            graph[a].append((b, 1))  # edge from a to b, original direction

            graph[b].append((a, 0))  # edge from b to a, reversed direction

        ## 3: [(1, 0), (2, 0)], => This means from 1 and 2, we can go to 3
        ## 4: [(0, 1), (5, 1)], => This means from 4, we can go to 0 and 5
        ## 0 means to that edge
        ## 1 means from that edge

        visited = set()
        queue = deque([0])
        res = 0

        queue


        while queue:
            current = queue.popleft() # Take the first city out of the queue to visit it.
            visited.add(current) # Mark the current city as visited, so we don’t process it again later.

            for neighbor, direction in graph[current]: # Look at all neighboring cities connected to the current city.
                if neighbor not in visited: #  We only want to process unvisited cities
                    res += direction #  If direction == 1, it means the road is going away from city 0, which is wrong, so we need to reverse it.
                    queue.append(neighbor)

        return res
                
```
