```python
def canReach(arr, start):
    visited = set()

    def dfs(i):
        # out of bounds
        if i < 0 or i >= len(arr):
            return False

        # already explored
        if i in visited:
            return False

        # found zero
        if arr[i] == 0:
            return True

        visited.add(i)

        return (
            dfs(i - arr[i]) or
            dfs(i + arr[i])
        )

    return dfs(start)
```
