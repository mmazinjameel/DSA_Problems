```python
from typing import List

class Solution:
    def findChampion(self, n: int, edges: List[List[int]]) -> int:
        in_degree = [0] * n

        for a, b in edges:
            in_degree[b] += 1

        champion = -1

        for i in range(n):
            if in_degree[i] == 0:
                if champion != -1:
                    return -1  # more than one candidate
                champion = i

        return champion
```
