```python
from typing import List

class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:
        
        rows = len(grid)
        cols = len(grid[0])
        visited = set()

        count = 0

        def helper(i,j):
            if i < 0 or j < 0 or i >= rows or j >= cols:
                return 1

            # If water → contributes to perimeter
            if grid[i][j] == 0:
                return 1

            # If already visited → no contribution
            if (i, j) in visited:
                return 0

            visited.add((i, j))

            # Explore all 4 directions
            return (
                helper(i + 1, j) +
                helper(i - 1, j) +
                helper(i, j + 1) +
                helper(i, j - 1)
            )

        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 1:   # only start from land
                    count += helper(i,j)

        return counts
```
