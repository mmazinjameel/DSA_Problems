# Recursion

```python
class Solution:
    def integerBreak(self, n: int) -> int:
        def dfs(num):
            if num == 1:
                return 1

            ans = 0

            for i in range(1, num):
                ans = max(
                    ans,
                    i * (num - i),   # stop breaking
                    i * dfs(num - i) # continue breaking
                )

            return ans

        return dfs(n)
```

# DP

```python
class Solution:
    def integerBreak(self, n: int) -> int:
        dp = [0] * (n + 1)

        dp[1] = 1

        for num in range(2, n + 1):
            for i in range(1, num):
                dp[num] = max(
                    dp[num],
                    i * (num - i),      # stop breaking
                    i * dp[num - i]     # continue breaking
                )

        return dp[n]
```
