# Recursion

```python
class Solution:
    def mincostTickets(self, days, costs):
        
        n = len(days)

        def dfs(i):

            # base case
            if i >= n:
                return 0

            # ---------- 1 day pass ----------
            cost1 = costs[0] + dfs(i + 1)

            # ---------- 7 day pass ----------
            j = i
            while j < n and days[j] < days[i] + 7:
                j += 1

            cost7 = costs[1] + dfs(j)

            # ---------- 30 day pass ----------
            j = i
            while j < n and days[j] < days[i] + 30:
                j += 1

            cost30 = costs[2] + dfs(j)

            return min(cost1, cost7, cost30)

        return dfs(0)
        
```

# DP

```python
class Solution:
    def mincostTickets(self, days: List[int], costs: List[int]) -> int:
        n = len(days)

        # dp[i] = minimum cost to cover travel days starting from days[i]
        dp = [0] * (n + 1)

        for i in range(n - 1, -1, -1):

            # 1-day pass
            cost1 = costs[0] + dp[i + 1]

            # 7-day pass
            j = i
            while j < n and days[j] < days[i] + 7:
                j += 1
            cost7 = costs[1] + dp[j]

            # 30-day pass
            j = i
            while j < n and days[j] < days[i] + 30:
                j += 1
            cost30 = costs[2] + dp[j]

            dp[i] = min(cost1, cost7, cost30)

        return dp[0]
```
