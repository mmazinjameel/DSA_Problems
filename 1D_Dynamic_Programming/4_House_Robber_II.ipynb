# Recursion

```python
from typing import List

class Solution:
    def rob(self, nums: List[int]) -> int:

        def dfs(arr, i):
            if i >= len(arr):
                return 0

            return max(dfs(arr, i + 1), arr[i] + dfs(arr, i + 2))

        if len(nums) == 1:
            return nums[0]

        # Case 1: exclude last house
        case1 = dfs(nums[:-1], 0)

        # Case 2: exclude first house
        case2 = dfs(nums[1:], 0)

        return max(case1, case2)
```

# Dynamic Programming

```python
from typing import List

class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)

        if n == 1:
            return nums[0]

        def helper(nums_):
            m = len(nums_)

            if m == 1:
                return nums_[0]

            dp = [0] * m
            dp[0] = nums_[0]
            dp[1] = max(nums_[0], nums_[1])

            for i in range(2, m):
                dp[i] = max(dp[i - 1], nums_[i] + dp[i - 2])

            return dp[-1]

        return max(helper(nums[1:]), helper(nums[:-1]))
```
