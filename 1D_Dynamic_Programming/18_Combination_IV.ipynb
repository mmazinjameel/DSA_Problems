# Recursion

```python
class Solution:
    def combinationSum4(self, nums: List[int], target: int) -> int:
        
        def backtrack(remain):
            if remain == 0:
                return 1

            if remain < 0:
                return 0

            total = 0

            for num in nums:
                total += backtrack(remain-num)

            return total

        return backtrack(target) 
```

# DP

```python
class Solution:
    def combinationSum4(self, nums: List[int], target: int) -> int:
        
        dp = [0] * (target + 1)
        dp[0] = 1

        for remain in range(1, target + 1):
            for num in nums:
                if remain - num >= 0:
                    dp[remain] += dp[remain - num]

        return dp[target]
```
