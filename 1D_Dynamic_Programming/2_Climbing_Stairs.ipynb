## Recursion

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        
        def solve(i):
            if i == n:
                return 1
            if i > n:
                return 0

            tw = solve(i + 2)
            on = solve(i + 1)

            return tw + on

        return solve(0)
```

# Dynamic Programming

```python
class Solution:
    def climbStairs(self, n: int) -> int:

        if n <= 2:
            return n

        dp = [0] * (n + 1)
        dp[1] = 1
        dp[2] = 2

        for i in range(3, n + 1):
            dp[i] = dp[i-1] + dp[i-2]

        return dp[n]
```
