# Recursion

```python
from collections import Counter

class Solution:
    def deleteAndEarn(self, nums):
        count = Counter(nums)
        unique = sorted(count.keys())

        n = len(unique)

        def dfs(i):

            if i >= n:
                return 0

            points = unique[i] * count[unique[i]]

            # skip current value
            skip = dfs(i + 1)

            # take current value
            if i + 1 < n and unique[i + 1] == unique[i] + 1:
                take = points + dfs(i + 2)
            else:
                take = points + dfs(i + 1)

            return max(take, skip)

        return dfs(0)
```

# DP


```python
from collections import Counter

class Solution:
    def deleteAndEarn(self, nums):
        count = Counter(nums)
        unique = sorted(count.keys())

        n = len(unique)

        dp = [0] * (n + 1)

        for i in range(n - 1, -1, -1):

            points = unique[i] * count[unique[i]]

            # skip current
            skip = dp[i + 1]

            # take current
            if i + 1 < n and unique[i + 1] == unique[i] + 1:

                take = points

                if i + 2 <= n:
                    take += dp[i + 2]

            else:
                take = points + dp[i + 1]

            dp[i] = max(take, skip)

        return dp[0]
```

# Forward DP

```python
from collections import Counter

class Solution:
    def deleteAndEarn(self, nums):
        count = Counter(nums)
        unique = sorted(count.keys())

        n = len(unique)

        dp = [0] * n

        dp[0] = unique[0] * count[unique[0]]

        for i in range(1, n):

            points = unique[i] * count[unique[i]]

            # consecutive values -> conflict
            if unique[i] == unique[i - 1] + 1:

                take = points
                if i > 1:
                    take += dp[i - 2]

                skip = dp[i - 1]

                dp[i] = max(take, skip)

            else:
                # no conflict
                dp[i] = dp[i - 1] + points

        return dp[-1]
```
