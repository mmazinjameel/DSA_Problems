```python
class Solution:
    def minimumIndex(self, nums: List[int]) -> int:
        
        # Moore Voting
        candidate = None
        count = 0

        for num in nums:
            if count == 0:
                candidate = num
            
            if num != candidate:
                count -= 1
            
            else:
                count += 1
        
        total = nums.count(candidate)

        left_count = 0
        n = len(nums)

        for i in range(n-1):
            if nums[i] == candidate:
                left_count += 1
            
            right_count = total - left_count

            left_len = i+1
            right_len = n - left_len

            if (left_count > left_len // 2 and right_count > right_len // 2):
                return i
        
        return -1

# At each index
# Move Current number to left side
# Reduce from right side
# Check if candidate dominates BOTH sides
```
