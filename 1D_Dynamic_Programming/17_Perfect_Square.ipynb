# Recursion

```python
class Solution:
    def numSquares(self, n: int) -> int:

        squares = []

        i = 1
        while i * i <= n:
            squares.append(i * i)
            i += 1
        

        def solve(idx, target):

            # target achieved
            if target == 0:
                return 0

            # no squares left
            if idx >= len(squares):
                return float('inf')

            take = float('inf')

            # take current square
            if squares[idx] <= target:
                take = 1 + solve(idx, target - squares[idx])

            # not take current square
            not_take = solve(idx + 1, target)

            return min(take, not_take)

        return solve(0, n)
```

# 2D DP

```python
class Solution:
    def numSquares(self, n: int) -> int:

        squares = []

        i = 1
        while i * i <= n:
            squares.append(i * i)
            i += 1

        dp = [[float("inf")] * (n + 1) for _ in range(len(squares) + 1)]

        for i in range(len(squares) + 1): # Because target 0 needs 0 squares
            dp[i][0] = 0

        for idx in range(len(squares) - 1, -1, -1):

            for target in range(1, n + 1):

                take = float("inf")

                if squares[idx] <= target:
                    take = 1 + dp[idx][target - squares[idx]]

                not_take = dp[idx + 1][target]

                dp[idx][target] = min(take, not_take)

        return dp[0][n]
```

# 1D DP

```python
class Solution:
    def numSquares(self, n: int) -> int:

        squares = []

        i = 1
        while i * i <= n:
            squares.append(i * i)
            i += 1

        dp = [float("inf")] * (n + 1)
        dp[0] = 0

        for target in range(1, n + 1):

            for square in squares:

                if square <= target:
                    dp[target] = min(
                        dp[target],
                        1 + dp[target - square]
                    )

        return dp[n]
```
