```python
class Solution:
    def subarraysDivByK(self, nums: List[int], k: int) -> int:
        count = 0
        prefix = {0:1}   
        curSum = 0

        for num in nums:
            curSum += num

            rem = curSum % k
            
            if rem in prefix:
                count += prefix[rem]
            
            prefix[rem] = prefix.get(rem, 0) + 1

        return count
    
# Core Idea:
# Remainder -> How many times we've seen it
# Difference between prefix sums is divisible by k
```
