```python
class Solution:
    def validPartition(self, nums):
        n = len(nums)

        # dp[i] means:
        # first i elements can be partitioned validly
        dp = [False] * (n + 1)

        # Empty array is valid
        dp[0] = True

        for i in range(2, n + 1):

            # Condition 1:
            # Last 2 elements are equal
            if nums[i - 1] == nums[i - 2]:
                dp[i] = dp[i] or dp[i - 2]

            # Conditions involving last 3 elements
            if i >= 3:

                # Condition 2:
                # Last 3 elements are equal
                if (nums[i - 1] == nums[i - 2] ==
                    nums[i - 3]):
                    dp[i] = dp[i] or dp[i - 3]

                # Condition 3:
                # Last 3 are consecutive
                if (nums[i - 3] + 1 == nums[i - 2] and
                    nums[i - 2] + 1 == nums[i - 1]):
                    dp[i] = dp[i] or dp[i - 3]

        return dp[n]
```
