```python
class Solution:
    def majorityElement(self, nums: List[int]) -> List[int]:
        freq = defaultdict(int)

        for num in nums:
            if num in freq:
                freq[num] += 1

            elif len(freq) < 2:
                freq[num] = 1

            else:
                for key in list(freq.keys()):
                    freq[key] -= 1
                    if freq[key] == 0:
                        del freq[key]
        res = []
        ## Extract the 2 elements
        for key in freq:
            if nums.count(key) > len(nums)//3:
                res.append(key)

        return res
```
