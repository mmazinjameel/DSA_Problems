# Recursion

```python
from typing import List

class Solution:
    def rob(self, nums: List[int]) -> int:

        def dfs(i):
            if i >= len(nums):
                return 0

            return max(dfs(i + 1), nums[i] + dfs(i + 2))

        return dfs(0)
```

# Dynamic Programming

```python
from typing import List

class Solution:
    def rob(self, nums: List[int]) -> int:
        if not nums:
            return 0
        if len(nums) == 1:
            return nums[0]

        n = len(nums)
        res = [0] * n

        res[0] = nums[0]
        res[1] = max(nums[0], nums[1])

        for i in range(2, n):
            res[i] = max(res[i - 1], nums[i] + res[i - 2])

        return res[-1]
```
