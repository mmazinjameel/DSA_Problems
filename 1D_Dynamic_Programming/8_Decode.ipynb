# Recursion

```python
class Solution:
    def numDecodings(self, s: str) -> int:
        def dfs(i):
        # reached the end → valid decoding
            if i == len(s):
                return 1
            
            # leading zero → invalid path
            if s[i] == '0':
                return 0
            
            # take single digit
            count = dfs(i + 1)
            
            # take two digits if valid (10–26)
            if i + 1 < len(s) and int(s[i:i+2]) <= 26:
                count += dfs(i + 2)
            
            return count
        
        return dfs(0)
```

# DP

```python
class Solution:
    def numDecodings(self, s: str) -> int:
        n = len(s)
        
        # dp[i] = number of ways to decode from index i to end
        dp = [0] * (n + 1)
        
        # base case: reaching the end = 1 way
        dp[n] = 1
        
        for i in range(n - 1, -1, -1):
            # if current is '0', no valid decoding
            if s[i] == '0':
                dp[i] = 0
                continue
            
            # take single digit
            dp[i] = dp[i + 1]
            
            # take two digits if valid
            if i + 1 < n and int(s[i:i+2]) <= 26:
                dp[i] += dp[i + 2]
        
        return dp[0]
```
