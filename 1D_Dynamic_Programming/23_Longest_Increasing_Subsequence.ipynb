# Recursion

```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        def dfs(i, prev):
            if i == len(nums):
                return 0

            skip = dfs(i + 1, prev)

            take = 0
            if prev == -1 or nums[i] > nums[prev]:
                take = 1 + dfs(i + 1, i)

            return max(take, skip)

        return dfs(0, -1)
```

# DP

```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        n = len(nums)

        dp = [[0] * (n + 1) for _ in range(n + 1)]

        for i in range(n - 1, -1, -1):
            for prev in range(i - 1, -2, -1):

                skip = dp[i + 1][prev + 1]

                take = 0
                if prev == -1 or nums[i] > nums[prev]:
                    take = 1 + dp[i + 1][i + 1]

                dp[i][prev + 1] = max(take, skip)

        return dp[0][0]
```
