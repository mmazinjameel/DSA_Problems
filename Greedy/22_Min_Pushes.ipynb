```python
class Solution:
    def minimumPushes(self, word: str) -> int:
        freq = [0] * 26

        # Count frequency of each character
        for ch in word:
            freq[ord(ch) - ord('a')] += 1

        # Sort in descending order
        freq.sort(reverse=True)

        pushes = 0

        # Assign push counts greedily
        for i in range(26):
            pushes += freq[i] * (i // 8 + 1)

        return pushes
```
