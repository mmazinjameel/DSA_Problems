```python
class Solution:
    def minimumSteps(self, s: str) -> int:

        # Keep track of how many 1s you’ve seen so far
        # Every time you see a 0, it must cross all previous 1s

        swaps = 0
        ones = 0

        for c in s:
            if c == '1':
                ones += 1
            else:  # c == '0'
                swaps += ones
        
        return swaps
```
