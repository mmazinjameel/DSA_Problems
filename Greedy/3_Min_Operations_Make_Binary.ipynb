```python
class Solution:
    def minOperations(self, nums: List[int]) -> int:
            n = len(nums)
            l = 0 # Left Pointer

            operations = 0

            for r in range(n):
                if r-l+1 == 3:
                    # If the leftmost element in the window is 0, flip all three
                    if nums[l] == 0:
                        for i in range(3):
                            nums[l+i] = 1 - nums[l+i]
                        operations += 1
                
                    # Move the left pointer 
                    l += 1
            # Check if all the elements are 1 after the operations
            if all(num==1 for num in nums):
                return operations
            else: return -1
```
