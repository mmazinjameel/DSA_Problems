```python
class Solution:
    def canBeValid(self, s: str, locked: str) -> bool:
        n = len(s)

        if n % 2 == 1:
            return False

        leftMin = 0
        leftMax = 0

        for i in range(n):

            if locked[i] == '1':

                if s[i] == '(':
                    leftMin += 1
                    leftMax += 1
                else:
                    leftMin -= 1
                    leftMax -= 1

            else:
                # can be '(' or ')'
                leftMin -= 1
                leftMax += 1

            if leftMax < 0:
                return False

            if leftMin < 0:
                leftMin = 0

        return leftMin == 0
```
