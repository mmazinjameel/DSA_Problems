```python
class Solution:
    def isNStraightHand(self, hand: List[int], groupSize: int) -> bool:
        if len(hand) % groupSize != 0:
            return False
        
        count = Counter(hand)
        min_heap = list(count.keys())
        heapq.heapify(min_heap)

        while min_heap:
            first = min_heap[0]

            for i in range(groupSize):
                current = first + i

                if current not in count:
                    return False

                count[current] -= 1

                if count[current] == 0: 
                    if current != min_heap[0]: # Because if current != min_heap[0] and count[current] == 0, then the heap is already inconsistent with the order required to build valid groups.
                        return False

                    heapq.heappop(min_heap)
                    del count[current]
        
        return True
```
