```python
from collections import Counter

class Solution:
    def canConstruct(self, s: str, k: int) -> bool:

        if len(s) < k:
            return False
        
        total_odds = sum(1 for odd in Counter(s).values() if odd % 2 != 0)

        if total_odds <= k:
            return True
        else:
            return False
```
