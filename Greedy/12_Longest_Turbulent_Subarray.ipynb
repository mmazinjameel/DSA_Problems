```python
class Solution:
    def maxTurbulenceSize(self, arr: List[int]) -> int:
        n = len(arr)

        if n == 1:
            return 1

        l = 0
        r = 1
        res = 1
        prev = ""

        while r < n:

            # current pair is decreasing
            if arr[r - 1] > arr[r] and prev != ">":
                prev = ">"
                res = max(res, r - l + 1)
                r += 1

            # current pair is increasing
            elif arr[r - 1] < arr[r] and prev != "<":
                prev = "<"
                res = max(res, r - l + 1)
                r += 1

            # turbulence breaks
            else:
                if arr[r] == arr[r - 1]:
                    r += 1

                l = r - 1
                prev = ""

        return res
```
