```python
class Solution:
    def maxAbsoluteSum(self, nums: List[int]) -> int:
        maxSum = 0
        minSum = 0

        currMax = 0
        currMin = 0

        for n in nums:
            currMax = max(n, currMax + n)
            maxSum = max(maxSum, currMax)

            currMin = min(n, currMin + n)
            minSum = min(minSum, currMin)

        return max(maxSum, abs(minSum))
```
