```python
class Solution:
    def maxScore(self, cardPoints: List[int], k: int) -> int:
        n = len(cardPoints)

        # Take first k cards initially
        curr = sum(cardPoints[:k])
        res = curr

        right = n - 1

        # Replace left cards with right cards one by one
        for left in range(k - 1, -1, -1):
            curr -= cardPoints[left]
            curr += cardPoints[right]
            right -= 1

            res = max(res, curr)

        return res
```
