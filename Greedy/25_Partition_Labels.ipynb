```python
class Solution:
    def partitionLabels(self, s: str) -> List[int]:

        dict_ = defaultdict(int)

        for i in range(len(s)):
            dict_[s[i]] = i


        max_index = 0
        res = []
        string = ""

        for i in range(len(s)):
            string += s[i]
            max_index = max(max_index, dict_[s[i]])

            if i == max_index:
                max_index = 0
                res.append(len(string))
                string = ""

        return res
```
