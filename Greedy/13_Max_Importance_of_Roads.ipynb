```python
class Solution:
    def maximumImportance(self, n: int, roads: List[List[int]]) -> int:
        array = [0] * n
        for i, j in roads:
            array[i] += 1
            array[j] += 1

        max_heap = []
        for v in array:
            heapq.heappush(max_heap, -v)

        total = 0
        value = n

        while max_heap:
            total += (-heapq.heappop(max_heap)) * value
            value -= 1

        return total

        
```
