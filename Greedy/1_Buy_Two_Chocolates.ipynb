```python
class Solution:
    def buyChoco(self, prices: List[int], money: int) -> int:
        prices.sort()
        left = money
        chocolates = 0
        
        for i in range(len(prices)):
            if left >= prices[i]:
                left -= prices[i]
                chocolates += 1
            
            if chocolates == 2:
                return left
        
        return money
```
