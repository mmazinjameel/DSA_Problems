```python
class Solution:
    def mergeNodes(self, head: ListNode) -> ListNode:
        p = head.next   # start after first 0
        q = p           # build result using q
        s = 0

        while p:
            if p.val != 0:
                s += p.val
            else:
                q.val = s   # store sum
                q.next = p.next
                q = q.next
                s = 0
            p = p.next

        return head.next
```
