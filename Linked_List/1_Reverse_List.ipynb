```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1:
            return list2
        if not list2:
            return list1
        
        p1 = list1
        p2 = list2

        # Initialize the merged list
        if p1.val < p2.val:
            third = last = p1.val
            p1 = p1.next
        else:
            third = last = ListNode(p2.val)
            p2 = p2.next

        # Merge the two lists
        while p1 and p2:
            if p1.val < p2.val:
                last.next = p1



            # Merge the two lists
            while p1 and p2:
                if p1.val < p2.val:
                    last.next = ListNode(p1.val)
                    p1 = p1.next
                else:
                    last.next = ListNode(p2.val)
                    p2 = p2.next
                last = last.next

            # Append remaining nodes
            while p1:
                last.next = ListNode(p1.val)
                p1 = p1.next
                last = last.next

            while p2:
                last.next = ListNode(p2.val)
                p2 = p2.next
                last = last.next

            return third



```
