```python
class Solution:
    def mergeInBetween(self, list1: ListNode, a: int, b: int, list2: ListNode) -> ListNode:
        p = q = list1
        temp1 = temp2 = list2   # fixed

        # Make q reach till b
        for i in range(b):
            q = q.next
        
        # Make p reach till a-1
        for i in range(a - 1):   # fixed
            p = p.next

        # Reach till end of list2
        while temp2.next != None:
            temp2 = temp2.next
        
        p.next = temp1
        temp2.next = q.next

        return list1
```
