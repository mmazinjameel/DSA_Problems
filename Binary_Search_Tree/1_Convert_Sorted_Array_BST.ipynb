```python
class Solution:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:

        def helper(l, hi):
            if l > hi:
                return None

            mid = (l + hi) // 2
            root = TreeNode(nums[mid])

            root.left = helper(l, mid - 1)
            root.right = helper(mid + 1, hi)

            return root

        return helper(0, len(nums) - 1)
```
