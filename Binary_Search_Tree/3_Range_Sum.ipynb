```python
class Solution:
    def rangeSumBST(self, root: Optional[TreeNode], low: int, high: int) -> int:

        range_sum = 0

        def dfs(root):
            nonlocal range_sum

            if not root:
                return

            if low <= root.val <= high:
                range_sum += root.val
            
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        return range_sum
```
