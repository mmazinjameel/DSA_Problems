```python
class Solution:
    def findMode(self, root: Optional[TreeNode]) -> List[int]:
        from collections import defaultdict
        
        if not root:
            return []
        
        freq = defaultdict(int)
        
        def dfs(node):
            if not node:
                return
            
            freq[node.val] += 1
            dfs(node.left)
            dfs(node.right)
        
        dfs(root)
        
        max_count = max(freq.values())
        
        return [key for key, value in freq.items() if value == max_count]
```

# O(1) - Space

```python
class Solution:
    def findMode(self, root: Optional[TreeNode]) -> List[int]:
        prev = None
        count = 0
        max_count = 0
        modes = []

        def inorder(node):
            nonlocal prev, count, max_count, modes

            if not node:
                return

            inorder(node.left)

            # Process current node
            if prev is not None and node.val == prev:
                count += 1
            else:
                count = 1

            if count > max_count:
                max_count = count
                modes = [node.val]   # reset list
            elif count == max_count:
                modes.append(node.val)

            prev = node.val

            inorder(node.right)

        inorder(root)
        return modes
```
