```python
import random
from typing import List

class Solution:

    def __init__(self, rects: List[List[int]]):
        self.rects = rects
        self.prefix = []
        total = 0
        
        for a, b, x, y in rects:
            # count integer points in this rectangle
            points = (x - a + 1) * (y - b + 1)
            total += points
            self.prefix.append(total)
        
        self.total_points = total

    def pick(self) -> List[int]:
        # Step 1: pick random number from 1 to total_points
        k = random.randint(1, self.total_points)
        
        # Step 2: binary search to find rectangle index
        left = 0
        right = len(self.prefix) - 1
        
        while left < right:
            mid = (left + right) // 2
            
            if self.prefix[mid] < k:
                # k is on the right side
                left = mid + 1
            else:
                # k could be at mid or on the left
                right = mid
        
        idx = left
        
        # Step 3: pick random point inside chosen rectangle
        a, b, x, y = self.rects[idx]
        
        rand_x = random.randint(a, x)
        rand_y = random.randint(b, y)
        
        return [rand_x, rand_y]
```
