```python
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        cur_sum = 0 
        minLenWindow = float('inf')

        i = 0
        j = 0

        while(j<len(nums)):
            cur_sum += nums[j]
            j+=1

            while(cur_sum>=target):

                minLenWindow = min(minLenWindow, j-i)

                cur_sum -= nums[i]
                i+=1

            

        return minLenWindow if minLenWindow != float('inf') else 0
        
```
