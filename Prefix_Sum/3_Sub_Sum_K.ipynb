```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        
        cur_sum = 0
        count = 0
        prefix = {0:1}

        for num in nums:
            cur_sum += num

            if cur_sum - k in prefix:
                count += prefix[cur_sum - k] # Add how many times I’ve seen a prefix sum equal to cur_sum - k before.

            prefix[cur_sum] = prefix.get(cur_sum,0) + 1

        return count
```
