```python
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        
        def same(a, b):
            if a is None and b is None:
                return True
            
            if a is None or b is None:
                return False
            
            if a.val != b.val:
                return False

            return same(a.left, b.left) and same(a.right, b.right) 
            

        return same(p,q)
```
