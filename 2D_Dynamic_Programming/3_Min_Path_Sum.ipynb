```python
class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        rows, cols = len(grid), len(grid[0])
        memo = {}

        def dfs(i, j):
            if i >= rows or j >= cols:
                return float('inf')

            # Reached destination
            if i == rows - 1 and j == cols - 1:
                return grid[i][j]

            if (i, j) in memo:
                return memo[(i, j)]

            # Move right and down
            right = dfs(i, j + 1)
            down = dfs(i + 1, j)

            memo[(i, j)] = grid[i][j] + min(right, down)
            return memo[(i, j)]

        return dfs(0, 0)
```
