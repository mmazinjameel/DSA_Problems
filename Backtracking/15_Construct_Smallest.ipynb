```python
class Solution:
    def smallestNumber(self, pattern: str) -> str:

        path = []

        used = [False] * 10
        ans = None

        def backtrack(i):

            nonlocal ans

            if ans is not None:
                return

            if i==len(pattern) + 1:
                ans = "".join(map(str,path))
                return

            for digit in range(1,10):

                if used[digit]:
                    continue
                
                if i>0:
                    prev = path[-1]

                    if pattern[i-1] == "I" and prev >= digit:
                        continue
                    if pattern[i-1] == "D" and prev <= digit:
                        continue
                
                path.append(digit)
                used[digit] = True

                backtrack(i+1)

                used[digit] = False
                path.pop()
        
        backtrack(0)
        return ans

            
```
