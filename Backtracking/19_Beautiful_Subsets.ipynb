```python
class Solution:
    def beautifulSubsets(self, nums: List[int], k: int) -> int:
        res = []

        def back(r, path):
            if r == len(nums):
                res.append(path[:])
                return

            res.append(path[:])

            for i in range(r, len(nums)):
                ok = True
                for x in path:
                    if abs(x - nums[i]) == k:
                        ok = False
                        break

                if ok:
                    path.append(nums[i])
                    back(i + 1, path)
                    path.pop()

        back(0, [])
        return len(res) - 1
```
