```python
class Solution:
    def findDifferentBinaryString(self, nums: List[str]) -> str:
        n = len(nums[0])
        nums = set(nums)

        def backtrack(curr):
            if len(curr) > n:
                return

            if len(curr) == n and curr not in nums:
                return curr

            res = backtrack(curr + "0")
            if res:
                return res

            res = backtrack(curr + "1")
            if res:
                return res

        return backtrack("")
```
