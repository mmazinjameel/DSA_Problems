```python
class Solution:
    def restoreIpAddresses(self, s: str) -> list[str]:
        res = []
        path = []

        def backtrack(i):
            # Valid IP found
            if i == len(s) and len(path) == 4:
                res.append(".".join(path))
                return

            # Invalid state
            if i == len(s) or len(path) == 4:
                return

            # Try taking 1, 2, or 3 digits
            for j in range(i, min(i + 3, len(s))):
                segment = s[i:j + 1]

                # Leading zero check
                if len(segment) > 1 and segment[0] == '0':
                    break

                # Range check
                if int(segment) > 255:
                    break

                path.append(segment)
                backtrack(j + 1)
                path.pop()

        backtrack(0)
        return res
```
