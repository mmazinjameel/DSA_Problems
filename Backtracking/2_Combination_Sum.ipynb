```python
from typing import List

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        res = []

        def backtracking(index, path, remaining):
            # If exact target achieved
            if remaining == 0:
                res.append(path[:])
                return
            
            # If out of bounds or exceeded target
            if index == len(candidates) or remaining < 0:
                return

            # Choice 1: skip current number
            backtracking(index + 1, path, remaining)

            if candidates[index] <= remaining:
            # Choice 2: include current number (stay at same index)
                path.append(candidates[index])
                backtracking(index, path, remaining - candidates[index])
                path.pop()

        backtracking(0, [], target)
        return res
```
