```python
class Solution:
    def splitString(self, s: str) -> bool:

        def backtrack(i, prev, count):
            if i == len(s):
                return count >= 2

            num = 0

            for j in range(i, len(s)):
                num = num * 10 + int(s[j])

                if prev is None:
                    if backtrack(j + 1, num, count + 1):
                        return True
                else:
                    if num == prev - 1:
                        if backtrack(j + 1, num, count + 1):
                            return True
                    elif num >= prev:
                        break

            return False

        return backtrack(0, None, 0)
```
