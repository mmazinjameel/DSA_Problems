```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        dict_ = {}

        for i in range(len(strs)):
            v = [0] * 26
            for j in range(len(strs[i])):
                v[ord(strs[i][j]) - ord('a')] += 1

            key = tuple(v)   # use tuple instead of set

            dict_.setdefault(key, []).append(strs[i])

        return list(dict_.values())
```
