```python
class Solution:
    def maxDistance(self, arrays: List[List[int]]) -> int:
        current_min = arrays[0][0]
        current_max = arrays[0][-1]
        difference = 0

        for i in range(1, len(arrays)):
            arr_min = arrays[i][0]
            arr_max = arrays[i][-1]

            difference = max(
                difference,
                abs(arr_max - current_min),
                abs(current_max - arr_min)
            )

            current_min = min(current_min, arr_min)
            current_max = max(current_max, arr_max)

        return difference
```
