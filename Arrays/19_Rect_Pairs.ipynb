```python
class Solution:
    def interchangeableRectangles(self, rectangles: List[List[int]]) -> int:
        
        ratios = {}
        count = 0

        for w, h in rectangles:

            ratio = w / h

            if ratio in ratios:
                count += ratios[ratio]

            ratios[ratio] = ratios.get(ratio, 0) + 1

        return count
```
