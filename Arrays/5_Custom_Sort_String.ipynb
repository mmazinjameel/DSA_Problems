```python
from collections import Counter

class Solution:
    def customSortString(self, order: str, s: str) -> str:
        count = Counter(s)
        res = ""

        # Add characters in the order specified
        for ch in order:
            if ch in count:
                res += ch * count[ch]
                del count[ch]

        # Add remaining characters
        for ch in count:
            res += ch * count[ch]

        return res
```
