```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:
        
        t = 0
        total = 0

        for arrival, order in customers:
            if t < arrival:  # Arriving after the idle time
                t = arrival
                t += order
                total += order
            else:  # covers both t == arrival and t > arrival
                total += t - arrival + order # Waiting time + Order making
                t += order

        return total / len(customers)
```
