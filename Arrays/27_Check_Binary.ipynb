```python
class Solution:
    def hasAllCodes(self, s: str, k: int):

        seen = set()

        for i in range(len(s) - k + 1):

            seen.add(s[i:i+k])

            if len(seen) == 2 ** k:
                return True

        return False

# Total possible binary strings of length k = 2^k
# If seen contains 2^k unique substrings, we have all codes
```
