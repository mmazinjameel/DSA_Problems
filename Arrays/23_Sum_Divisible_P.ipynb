```python
class Solution:
  def minSubarray(self, nums: list[int], p: int) -> int:
    summ = sum(nums)
    remainder = summ % p
    if remainder == 0:
      return 0

    ans = len(nums)
    prefix = 0
    prefixToIndex = {0: -1} # Previous Prefix : index - i

    for i, num in enumerate(nums):
      prefix = (prefix + num) % p
    
      needed_prefix = (prefix - remainder + p) % p
      if needed_prefix in prefixToIndex:
        ans = min(ans, i - prefixToIndex[needed_prefix])
      prefixToIndex[prefix] = i

    return -1 if ans == len(nums) else ans

# Suppose sum % p = x
# Need to remove y from sum to make remainder 0: (sum-y)%p=x
# Therefore, y%p = x. That is, find the subarray whose sum is such that the remainder is x (whole array's sum remainder)
# Now, subarray % p = remainder
# subarray = currentPrefix − previousPrefix
# (currentPrefix − previousPrefix) % p = remainder
# previousPrefix % p = (currentPrefix − remainder) % p
# We need to find previousPrefix in the hashmap, and if it's there, take it's starting index
```
