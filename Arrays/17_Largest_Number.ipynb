```python
class Solution:
    def largestNumber(self, nums):

        nums = [str(x) for x in nums]

        n = len(nums)

        # bubble sort
        for i in range(n):

            for j in range(i + 1, n):

                # decide order
                if nums[j] + nums[i] > nums[i] + nums[j]:

                    nums[i], nums[j] = nums[j], nums[i]

        answer = ''.join(nums)

        # case like [0,0]
        if answer[0] == '0':
            return '0'

        return answer
```
