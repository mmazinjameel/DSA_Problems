# Recursion

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        
        def dfs(i, holding):

            # base case
            if i == len(prices):
                return 0
            
            # If we don't have any
            if not holding:
                buy = -prices[i] + dfs(i+1, True)

                skip = dfs(i+1, False)

                return max(buy,skip)
            
            elif holding:
                sell = prices[i] + dfs(i+1, False)
                skip = dfs(i+1,True)

                return max(sell, skip)
        
        return dfs(0,False)
```

# 2D DP

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:

        n = len(prices)

        # Create Array
        dp = [[0] * 2 for _ in range(n+1)]

        # Build Backwards
        for i in range(n-1,-1,-1):

            # not holding
            dp[i][0] = max(-prices[i] + dp[i+1][1], dp[i+1][0])

            # holding
            dp[i][1] = max(prices[i] + dp[i+1][0], dp[i+1][1])

        return dp[0][0]
        
```

# 1D DP

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:

        n = len(prices)

        ahead = [0, 0]

        for i in range(n - 1, -1, -1):

            cur = [0, 0]

            # Not Holding
            cur[0] = max(-prices[i] + ahead[1], ahead[0])

            # Holding
            cur[1] = max(prices[i] + ahead[0], ahead[1])

            ahead = cur

        return ahead[0]
```
