```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        
        v1 = [0] * 26
        v2 = [0] * 26
        res = []

        for i in range(len(p)):
            v1[ord(p[i]) - ord('a')] += 1

        left = 0

        for right in range(len(s)):

            v2[ord(s[right]) - ord('a')] += 1

            # window size exceeds len(p)
            if right - left + 1 > len(p):
                v2[ord(s[left]) - ord('a')] -= 1
                left += 1

            # compare both frequency arrays
            if right - left + 1 == len(p):
                if v1 == v2:
                    res.append(left)

        return res
```
