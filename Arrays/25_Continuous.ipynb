```python
class Solution:
    def checkSubarraySum(self, nums: list[int], k: int) -> bool:

        prefix = 0

        # remainder -> earliest index
        prefixToIndex = {0: -1}

        for i, num in enumerate(nums):

            prefix = (prefix + num) % k

            # same remainder found before
            if prefix in prefixToIndex:

                # subarray length must be at least 2
                if i - prefixToIndex[prefix] >= 2:
                    return True

            else:
                # store first occurrence only
                prefixToIndex[prefix] = i

        return False
```
