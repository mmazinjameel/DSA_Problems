```python
class Solution:
    def vowelStrings(self, words: List[str], queries: List[List[int]]) -> List[int]:

        vowel = ('a','e','i','o','u')
        
        hashmap_ = {}

        hashmap_[0] = 1 if (words[0][0] in vowel and words[0][-1] in vowel) else 0

        for i in range(1,len(words)):
            if words[i][0] in vowel and words[i][-1] in vowel:
                hashmap_[i] = hashmap_[i-1] + 1 
            else:
                hashmap_[i] = hashmap_[i-1]

        res = []

        for x,y in queries:
            if words[x][0] in vowel and words[x][-1] in vowel:
                res.append(hashmap_[y]-hashmap_[x]+1)
            else:
                res.append(hashmap_[y]-hashmap_[x])

        return res
```
