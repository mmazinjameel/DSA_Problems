```python
class Solution:
    def minSwaps(self, s: str) -> int:
        
        s = list(s)
        r = len(s) - 1
        bal = swaps = 0

        for i in range(len(s)):

            bal += 1 if s[i] == '[' else -1

            if bal < 0:

                while s[r] != '[':
                    r -= 1

                s[i], s[r] = s[r], s[i]
                swaps += 1
                bal = 1
                r -= 1

        return swaps
```
