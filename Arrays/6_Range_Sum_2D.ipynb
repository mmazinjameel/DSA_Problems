```python
class NumMatrix:

    def __init__(self, matrix: List[List[int]]):

        if not matrix or not matrix[0]:
            self.prefix = []
            return

        rows, cols = len(matrix), len(matrix[0])

        # Create prefix matrix
        self.prefix = [[0] * cols for _ in range(rows)]

        # Build row-wise cumulative sum
        for r in range(rows):
            running_sum = 0
            for c in range(cols):
                running_sum += matrix[r][c]
                self.prefix[r][c] = running_sum

    def sumRegion(self, row1: int, col1: int, row2: int, col2: int) -> int:
        total = 0

        for r in range(row1, row2 + 1):
            if col1 > 0:
                total += self.prefix[r][col2] - self.prefix[r][col1 - 1]
            else:
                total += self.prefix[r][col2]

        return total
        


# Your NumMatrix object will be instantiated and called as such:
# obj = NumMatrix(matrix)
# param_1 = obj.sumRegion(row1,col1,row2,col2)
```
