```python
class Solution:
    def gridGame(self, grid: List[List[int]]) -> int:
        
        topSum = sum(grid[0])
        bottomSum = 0
        minSum = float('inf')

        for partition in range(len(grid[0])):

            topSum -= grid[0][partition]

            minSum = min(minSum,
                         max(topSum, bottomSum))

            bottomSum += grid[1][partition]

        return minSum
```
