```python
class Solution:
    def numOfSubarrays(self, arr: List[int]) -> int:
    
        count = 0
        prefix = {0:1}
        curSum = 0
        MOD = 10**9 + 7

        for num in arr:
            curSum += num

            rem = curSum % 2
            
            if 1 - rem in prefix:
                count += prefix[1 - rem]
            
            prefix[rem] = prefix.get(rem, 0) + 1

        return count % MOD
```
